```python
import pandas as pd
import selenium
from selenium import webdriver 
from selenium.webdriver.support.ui import Select
import time
from itertools import zip_longest
driverPath=r"C:\Users\Merit\Desktop\Pk\Day 10 - 13,15 - 19,21-24 (PYTHON)\chromedriver_win32\chromedriver.exe"
driver = webdriver.Chrome(executable_path = driverPath)
driver.maximize_window()
driver.get("https://www.amazon.in/s?k=offers+on+lipstick&ref=nb_sb_noss_1")

writer = pd.ExcelWriter('Assignment-4.xlsx',engine = 'xlsxwriter')

lipstick = driver.find_elements_by_xpath("//h2[contains(@class,'s-line-clamp-4')]")
old_price = driver.find_elements_by_xpath("//span[contains(@class,'a-text-price')]")
offer_price = driver.find_elements_by_xpath("//span[contains(@class,'a-price-whole')]")

li = []
li1 = []
li2 = []
for i in lipstick:
    li.append(i.text)
print(li)

print("----------------------------")
for j in old_price:
    li1.append(j.text)
print(li1)

print("----------------------------")

for k in offer_price:
    li2.append(k.text)
print(li2)    
print("----------------------------")

l=zip_longest(li,li1,li2)

df = pd.DataFrame(l,columns=["Name","old_price","offer_price"])
df.to_excel(writer,sheet_name = 'lipstick' ,index = False)
writer.save()
df
```
